```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
dw_file = "Resources/dw.csv"
dw_df = pd.read_csv(dw_file)
dw_df.head()
```

```python
drinks_file = "Resources/drinks.csv"
drinks_df = pd.read_csv(drinks_file)
drinks_df.head()
```

```python
dw2019_df = dw_df.loc[dw_df['Year'] > 2018]
dw2019_df.head()
```

```python
dw_cols = ['Country', 'Score', 'GDP', 'Social_Support', 'Freedom_to_choices', 'Perceptions_of_corruption']
dw_transformed = dw2019_df[dw_cols].copy()

dw_transformed = dw_transformed.rename(columns={'Country': 'country',
                                               'Score': 'happiness_score',
                                               'GDP': 'gdp',
                                               'Social_Support': 'social_support',
                                               'Freedom_to_choices': 'freedom_of_choice',
                                               'Perceptions_of_corruption': 'perception_of_corruption'})

dw_transformed.set_index('country', inplace=True)

dw_transformed.head()
```

```python
# country	beer_servings	spirit_servings	wine_servings	total_litres_of_pure_alcohol
drinks_cols = ['country', 'beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']
drinks_transformed = drinks_df[drinks_cols].copy()

drinks_transformed.set_index('country', inplace=True)
drinks_transformed.head()
```

```python
connection_string = "postgres:postgres@localhost:5432/happiness_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

```python
dw_transformed.to_sql(name ='dw', con=engine, if_exists='append', index=True)
```

```python
drinks_transformed.to_sql(name ='drinks', con=engine, if_exists='append', index=True)
```
